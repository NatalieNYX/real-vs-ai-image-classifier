Import Function 

```python
#class that loads and preprocesses images using keras/tensorflow
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

Normalization + Augmentation

```python
#create image generator 
#as images now are in .jpg, with pixel value of 0-255
#normalize to 0-1

image_datagen = ImageDataGenerator(
    rescale = 1./255, # normalization 
)

#create training image generator with augmentation applied
train_datagen = ImageDataGenerator(
    rescale = 1./255,
    rotation_range = 20, # randomly rotate images in the range (degrees, 0 to 180)
    width_shift_range =0.1, # randomly shift images horizontally (fraction of total width)
    height_shift_range = 0.1, # randomly shift images vertically (fraction of total height)
    zoom_range = 0.1, # randomly zoom images
    horizontal_flip = True # randomly flip images horizontally
)
```

Image Loading 

```python
train_dir = "../dataset/train"
valid_dir = "../dataset/valid"
test_dir = "../dataset/test"

#flow_from_directory -> load images directly from folder structure and convert to batches of tensors
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size = (128, 128), #make all image same size (pixel)
    batch_size = 32, # 32 images can be processed at once, adjust based on memory
    class_mode = "binary" # type of label the generator will produce
)

valid_generator = image_datagen.flow_from_directory(
    valid_dir,
    target_size = (128, 128), 
    batch_size = 32,
    class_mode = "binary"
)

test_generator = image_datagen.flow_from_directory(
    test_dir,
    target_size = (128, 128),
    batch_size = 32,
    class_mode = "binary",
    shuffle = False #shuffle true only for training data for randomization
)
```
